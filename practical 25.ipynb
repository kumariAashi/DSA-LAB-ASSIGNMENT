```python
# Dynamic Programming - Fibonacci

def fibonacci(n):
    dp = [0] * (n + 1)

    if n >= 1:
        dp[1] = 1

    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]

    return dp[n]

# Main Program
n = int(input("Enter a number: "))

print("Fibonacci number is:", fibonacci(n))
```
