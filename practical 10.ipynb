```python
#Row-Major order representation
def demonstrate_row_major():
    matrix = [
        [8, 10],
        [11, 13],
        [14, 16]
    ]

    rows = 3
    cols = 2

    flat_memory = []

    print("Logical 2D Matrix View:")
    for r in matrix:
        print(r)
        for element in r:
            flat_memory.append(element)

    print("\nRow-Major Linear Representation in Memory:")
    print(flat_memory)


# Function call
demonstrate_row_major()
```
